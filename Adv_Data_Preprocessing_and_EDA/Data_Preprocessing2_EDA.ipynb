```python
# How does income level influence purchasing patterns across different product categories?
```

```python
df["Income"].describe()
```

```python
def incomeComparator(Income):
    if (Income <= 35397.000000):
        return 'Low Income'
    elif (Income > 35397.000000 and Income < 52161.539133):
        return 'Below Average Income'
    elif (Income >= 52161.539133 and Income <= 68655.500000):
        return 'Above Average Income'
    else :
        return 'High Income'
```

```python
df["Income_Category"] = df["Income"].apply(incomeComparator)
```

```python
df.head()
```

```python
df_income = df.groupby('Income_Category')[['MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts', 'MntGoldProds']].sum()
df_income
```

```python
df_income['Income_Category'] = df_income.index
```

```python
df_income
```

```python
df_income[df_income['Income_Category']=='Low Income'][['MntWines' , 'MntFruits' , 'MntMeatProducts' , 'MntFishProducts'	,'MntSweetProducts',	'MntGoldProds']].to_numpy().tolist()[0]
```

```python
categories = ['Low Income', 'Below Average Income', 'Above Average Income', 'High Income']
labels = ['MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts', 'MntGoldProds']


for i in categories:
    expend = df_income[df_income['Income_Category'] == i][['MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts', 'MntGoldProds']].to_numpy().tolist()[0]

    plt.figure(figsize=(7, 7))
    plt.pie(expend, labels=labels, colors=['brown', 'blue', 'green', 'red', 'pink', 'yellow'], autopct='%1.1f%%', startangle=140)

    plt.title(f"Expenditure Distribution for {i}", fontsize=16, fontweight='bold')
    plt.axis('equal')
    plt.tight_layout()
    plt.show()
```

```python
# Do households with children/teenagers show different spending patterns compared to those without?
```

```python
def check_fam (Kidhome,Teenhome) :
  if(Teenhome > 0 and Kidhome > 0) :
    return 'Family with Teen and Kid'
  elif(Kidhome > 0) :
    return 'Family With Kid'
  elif(Teenhome > 0) :
    return 'Family with Teen'
  else :
    return 'Single'
```

```python
df['Family_Status'] = df.apply(lambda x: check_fam(x['Kidhome'], x['Teenhome']), axis=1)
```

```python
df.groupby('Family_Status')['Total_Spending'].sum().plot(kind='bar')
```
