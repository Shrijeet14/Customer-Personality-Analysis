We Can also see clear issues in the Teenhome column as we can see it has many dates values in it and the the coun of Teens.

```python
df['Teenhome'].unique()
```

We Can notice that the Teenhome column has only 0  , 1 , 2 as valid data , others are just dates which is not correctwe need to deal with this issue.

```python
df[df['Teenhome'].isin(['1', '2', '0'])]
```

```python
df[~df['Teenhome'].isin(['1', '2', '0'])]
```

To deal with this issue we wil simply replace the values with mode of the column as it is a categorical feature.

```python
df.loc[~df['Teenhome'].isin(['1', '2', '0']) , 'Teenhome'] = df['Teenhome'].mode()[0]
```

```python
# Let's Check the values of some of those rows now
df.iloc[[10 , 26 , 41 , 45]]
```

We can see the issue is resolved now .
